```python
def display_gamelist(mdict):
    print('This is the current game board:')
    print(str(mdict['p1'])+'  |  '+str(mdict['p2'])+'  |  '+str(mdict['p3'])+'  |  ')
    print(str(mdict['p4'])+'  |  '+str(mdict['p5'])+'  |  '+str(mdict['p6'])+'  |  ')
    print(str(mdict['p7'])+'  |  '+str(mdict['p8'])+'  |  '+str(mdict['p9'])+'  |  ')
```

```python
def position_choice(picked_positions):
    choice='wrong'
    while choice not in range(1,10):
        choice=input('Please enter a position in the range (1,9)')
        if choice.isdigit()==False:
            print('The position entered is not a digit, please try again!')
        elif int(choice) not in range(1,10):
            print('The position entered is not in the range(1,9)')
        elif int(choice) in picked_positions:
            print('The position entered has already been picked, please pick a different one')
        else:
            choice=int(choice)
    print(choice)
    return choice
```

```python
def replacement(mdict,position,symbolp):
    mdict['p'+str(position)]=symbolp
    return mdict
```

```python
def gameon():
    choice='wrong'
    while choice not in ['Y','N']:
        choice=input('Would you like to continue game (Y,N)?')
        if choice not in ['Y','N']:
            print('Please try again')
    if choice=='Y':
        return True
    else:
        return False
```

```python
def empty_str(x):
    if x==' ':
        return True
    else:
        return False
```

```python
def board_full(x):
    mark=['O','X']
    if x in mark:
        return True
    else:
        return False
```

```python
def check_diag(board):
    #Check Diagonals
    check_set=set(board[n][n] for n in range(len(board)))
    if len(set(list(filter(empty_str,check_set))))==1:
        return False
    elif len(set(board[n][n] for n in range(len(board))))==1:
        return True
    else:
        return False
```

```python
def check_rows(board):
    #Check rows
    for r in board:
        if len(set(r))==1 and set(r)!={' '}:
            return True
        else:
            return False
```
